```r
data <- read.csv("C:/Users/Sandeepa/Music/heart.csv")
#print(data)
boxplot(MaxHR ~ Sex, 
        data = data, 
        main = "Maximum Heart Rate", 
        xlab = "Sex", 
        ylab = "Maximum Heart Rate", 
        col = c("purple", "yellow"))

outliers_male <- boxplot.stats(data$MaxHR[data$Sex == "M"])$out
outliers_female <- boxplot.stats(data$MaxHR[data$Sex == "F"])$out

cat("Outliers for Males :", outliers_male, "\n")
cat("Outliers for Females :", outliers_female, "\n")

install.packages("moments")
library(moments)

skewness_male <- skewness(data$MaxHR[data$Sex == "M"], na.rm = TRUE)
skewness_female <- skewness(data$MaxHR[data$Sex == "F"], na.rm = TRUE)

cat("Skewness for Males :", skewness_male, "\n")
cat("Skewness for Females :", skewness_female, "\n")
```

```r
data2 <- read.csv("C:/Users/Sandeepa/Music/games.csv")
#print(head(data, 10))
boxplot(turns ~ winner, 
        data = data2, 
        xlab = "Winner of the game", 
        ylab = "No of turns", 
        main = "Chess game summary", 
        col = c("green", "blue", "red"))

outliers_white <- boxplot.stats(data2$turns[data2$winner == "white"])$out
outliers_black <- boxplot.stats(data2$turns[data2$winner == "black"])$out
outliers_draw <- boxplot.stats(data2$turns[data2$winner == "draw"])$out

cat("Outliers for White (Winner = white):", outliers_white, "\n")
cat("Outliers for Black (Winner = black):", outliers_black, "\n")
cat("Outliers for Draw (Winner = draw):", outliers_draw, "\n")

install.packages("moments")
library(moments)
skewness_white <- skewness(data2$turns[data2$winner == "white"], na.rm = TRUE)
skewness_black <- skewness(data2$turns[data2$winner == "black"], na.rm = TRUE)
skewness_draw <- skewness(data2$turns[data2$winner == "draw"], na.rm = TRUE)

cat("Skewness for White:", skewness_white, "\n")
cat("Skewness for Black:", skewness_black, "\n")
cat("Skewness for Draw:", skewness_draw, "\n")

set.seed(123)
mydata1 <- rnorm(120, mean = 50, sd = 12)
boxplot(mydata1, 
        main = "Random Data Distribution",
        ylab = "Values",
        col = "cyan")

five_number_summary <- summary(mydata1)
min_value <- min(mydata1, na.rm = TRUE)
max_value <- max(mydata1, na.rm = TRUE)
first_quartile <- quantile(mydata1, 0.25, na.rm = TRUE)
third_quartile <- quantile(mydata1, 0.75, na.rm = TRUE)
median_value <- median(mydata1, na.rm = TRUE)
range_value <- range(mydata1)
skewness_value <- skewness(mydata1, na.rm = TRUE)
outliers_mydata1 <- boxplot.stats(mydata1)$out

cat("Five Number Summary:", five_number_summary, "\n")
cat("Minimum:", min_value, "\n")
cat("Maximum:", max_value, "\n")
cat("First Quartile (Q1):", first_quartile, "\n")
cat("Third Quartile (Q3):", third_quartile, "\n")
cat("Median:", median_value, "\n")
cat("\nRange:", range_value, "\n")
cat("\nSkewness:", skewness_value, "\n")
cat("\nOutliers:", outliers_mydata1, "\n")
```
